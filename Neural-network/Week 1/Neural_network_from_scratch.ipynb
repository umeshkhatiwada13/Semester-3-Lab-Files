# Perceptron

```python
import random
```

```python
# Initializing empty lists to store input and output data
input_data = []
output_data = []

# Generating data for 100 samples
for i in range(1, 100):
    # Generating random float numbers between 0 and 0.5 for inputs
    input1 = round(random.uniform(0, 0.5), 2)
    input2 = round(random.uniform(0, 0.5), 2)
    
    # Determining the output based on the sum of inputs
    # If the sum of inputs is less than 0.5, output is 0; otherwise, output is 1
    output_1 = 0 if input1 + input2 < 0.5 else 1
    
    # Appending the input and output data to the respective lists
    input_data.append([input1, input2])
    output_data.append(output_1)

# Printing the generated input and output data
print(input_data)
print(output_data)
```

```python
# initialize weights
w1 = random.uniform(-1, 1)
w2 = random.uniform(-1, 1)
b = random.uniform(-1, 1)
w1, w2, b
```

```python
# Learning rate
lr = 0.01
# Number of trials
epochs = 100
```

```python
# train test split : 90 % for training and 10% for testing
train_size = int(0.9 * len(input_data))
print(train_size)
```

```python
# Define activation function
def activate(weighted_sum):
    threshold = 0
    if weighted_sum >= threshold:
        return 1
    else:
        return 0
```

```python
for epoch in range(epochs):
    total_error = 0
    print("\nEpoch: ", epoch)

    for i in range(len(input_data[:train_size])):
        x1, x2 = input_data[i]
        y = output_data[i]
        print("Data: ", x1, x2, ", Expected output: ", y)

        # Calculate weighted sum
        weighted_sum = w1 * x1 + w2 * x2 + b
        print("weighted_sum: ", weighted_sum)
        output = activate(weighted_sum)
        print("Predicted output: ", output)
        error = y - output
        print("Error: ", error)

        # Update weights
        w1 += lr * error * x1
        w2 += lr * error * x2
        b += lr * error

        total_error += abs(error)
        print("total_error: ", total_error)

    if total_error == 0:
        print("\n Model learned at epoch: ", epoch)
        break

```

```python
# Learned weights
print("Learned weights(w1, w2, b): ", w1, w2, b)
```

```python
# OR gate test dataset
test_data = input_data[train_size:]
expected_test_output = output_data[train_size:]
test_data, expected_test_output
```

```python
# Initialize a variable to count the number of accurate predictions
accurate_prediction = 0

# Loop through the test data
for i in range(len(test_data)):
    # Extract the input values (x1 and x2) from the test data
    x1, x2 = test_data[i]
    
    # Calculate the weighted sum of inputs using the trained weights (w1 and w2) and bias (b)
    weighted_sum = w1 * x1 + w2 * x2 + b
    
    # Apply the activation function to get the predicted output
    predicted_test_output = activate(weighted_sum)
    
    # Compare the predicted output with the expected output from the test data
    if predicted_test_output == expected_test_output[i]:
        # If the prediction matches the expected output, increment the count of accurate predictions
        accurate_prediction += 1
    
    # Print the input values, predicted output, and expected output for each test sample
    print(
        f"Input: ({x1}, {x2}) - Predicted Output: {predicted_test_output}; Expected Output: {expected_test_output[i]}")
```

```python
# Calculate the testing accuracy by dividing the number of accurate predictions by the total number of samples in the test dataset
testing_accuracy = accurate_prediction / len(test_data)

# Convert the testing accuracy to a percentage by multiplying it by 100
testing_accuracy_percentage = testing_accuracy * 100

# Print the testing accuracy percentage
print(f"Testing Accuracy: {testing_accuracy_percentage:.2f} %")
```
