```python
import os
import pygame
from deepgram import DeepgramClient, SpeakOptions
from groq import Groq
from datetime import datetime

# Initialize pygame mixer
pygame.mixer.init()

# Define API keys and options
DEEPGRAM_API_KEY = "b7abf9d8190a25b5f93ca15ff8f5d048df0987bd"  # Replace with your actual Deepgram API key
GROQ_API_KEY = "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"  # Replace with your actual Groq API key
output_dir = "audio_output"  # Directory to save audio files

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Initialize Deepgram client
deepgram = DeepgramClient(api_key=DEEPGRAM_API_KEY)

# Initialize Groq client
groq_client = Groq(api_key=GROQ_API_KEY)

def get_groq_response(user_input, conversation_history):
    # Add user input to conversation history
    conversation_history.append({
        "role": "user",
        "content": user_input
    })

    # Generate response from the model
    chat_completion = groq_client.chat.completions.create(
        messages=conversation_history,
        model="llama3-8b-8192"
    )

    ai_response = chat_completion.choices[0].message.content

    # Add AI response to conversation history
    conversation_history.append({
        "role": "assistant",
        "content": ai_response
    })

    return ai_response, conversation_history

def save_to_file(filename, text):
    # Generate unique filename with timestamp
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = os.path.join(output_dir, f"{filename}_{timestamp}.mp3")

    # Configure Deepgram options
    options = SpeakOptions(model="aura-asteria-en")

    # Save the audio content to the file
    response = deepgram.speak.v("1").save(filename, {"text": text}, options)

    return filename

def play_audio(filename):
    pygame.mixer.music.load(filename)
    pygame.mixer.music.play()

    # Wait for the audio to finish playing
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

def main():
    try:
        # Initial system message setting the context
        conversation_history = [
            {
                "role": "system",
                "content": "You are a receptionist at a Heating and Ventilation and Air Conditioning Company. Be resourceful and efficient. Your Company sells Heat Pumps and Air Conditioners. The available sales agents are Raju and Sanat. Only answer in two sentences."
            }
        ]

        while True:
            user_input = input("User: ")

            # Get Groq response
            ai_response, conversation_history = get_groq_response(user_input, conversation_history)
            print("AI:", ai_response)

            # Save AI response to file
            filename = save_to_file("output", ai_response)
            print(f"File '{filename}' saved successfully.")

            # Play the saved audio file using pygame
            play_audio(filename)

            # Exit condition example
            if user_input.lower() == "exit":
                break

    except Exception as e:
        print(f"Exception: {e}")

if __name__ == "__main__":
    main()
```

```python

# Define API keys and options
DEEPGRAM_API_KEY = "b7abf9d8190a25b5f93ca15ff8f5d048df0987bd"  # Replace with your actual Deepgram API key
# GROQ_API_KEY = "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"  # Replace with your actual Groq API key
output_dir = "audio_output"  # Directory to save audio files

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Initialize Deepgram client
deepgram = DeepgramClient(api_key=DEEPGRAM_API_KEY)

# Function to handle transcription received from Deepgram
async def handle_transcript(transcript):
    if 'channel' in transcript:
        alternatives = transcript['channel']['alternatives']
        if alternatives and 'transcript' in alternatives[0]:
            print("Transcript:", alternatives[0]['transcript'])

# Function to stream audio to Deepgram
async def stream_audio():
    chunk = 1024  # Record in chunks of 1024 samples
    sample_format = pyaudio.paInt16  # 16 bits per sample
    channels = 1
    rate = 44100  # Record at 44100 samples per second

    p = pyaudio.PyAudio()  # Create an interface to PortAudio

    print("Recording and streaming...")

    stream = p.open(format=sample_format,
                    channels=channels,
                    rate=rate,
                    frames_per_buffer=chunk,
                    input=True)

    async def generator():
        while True:
            data = stream.read(chunk)
            yield data

    options = {
        "punctuate": True,
        "interim_results": False,
        "language": "en-US"
    }

    async with deepgram.transcription.live(options=options) as conn:
        async for data in generator():
            await conn.send(data)
            async for transcript in conn:
                await handle_transcript(transcript)

    stream.stop_stream()
    stream.close()
    p.terminate()

# Main entry point
async def main():
    try:
        await stream_audio()
    except KeyboardInterrupt:
        pass
    finally:
        await asyncio.sleep(1)  # Wait for any remaining tasks to finish
        asyncio.get_event_loop().close()

if __name__ == "__main__":
    asyncio.run(main())
```

```python
import pyaudio
from deepgram import DeepgramClient, LiveTranscriptionEvents, LiveOptions

# Constants for audio input
sample_format = pyaudio.paInt16
channels = 1
rate = 44100
chunk = 1024
seconds_to_record = 10  # Duration of audio to record in seconds

# Initialize PyAudio
p = pyaudio.PyAudio()

# Open audio stream
stream = p.open(format=sample_format,
                channels=channels,
                rate=rate,
                frames_per_buffer=chunk,
                input=True)

# Define API keys and options
DEEPGRAM_API_KEY = "b7abf9d8190a25b5f93ca15ff8f5d048df0987bd"  #

# Create a websocket connection using the DEEPGRAM_API_KEY from environment variables
deepgram = DeepgramClient(DEEPGRAM_API_KEY)

# Use the listen.live class to create the websocket connection
dg_connection = deepgram.listen.live.v("1") 

# Start the WebSocket connection with options
options = LiveOptions(
  punctuate=True,
  interim_results=False,
  language='en-GB'
)
dg_connection.start(options)

# Function to generate audio data chunks
def generate_audio_chunks(duration):
    frames = []
    num_frames = int(rate / chunk * duration)
    print(num_frames)
    for i in range(num_frames):
        data = stream.read(chunk)
        frames.append(data)
    return frames

# Generate and send audio data chunks
audio_chunks = generate_audio_chunks(seconds_to_record)
print("Audio Chunk",len(audio_chunks))
for chunk_data in audio_chunks:
    dg_connection.send(chunk_data)

# Event handler for transcription results
def on_message(result, **kwargs):
    sentence = result.channel.alternatives[0].transcript
    print(sentence)
    if len(sentence) == 0:
        return
    print(f"Transcription: {sentence}")

# Register event handler for transcription events
dg_connection.on(LiveTranscriptionEvents.Transcript, on_message)

# Finish WebSocket connection
dg_connection.finish()

# Clean up PyAudio stream and close
stream.stop_stream()
stream.close()
p.terminate()
```
